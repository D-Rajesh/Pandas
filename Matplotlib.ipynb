```python
x=['Python','C','c++','Java']
y=[85,60,65,79]
z=[45,36,25,65]
plt.xlabel('Language',fontsize=15)
plt.ylabel('Bhosodi',fontsize=15)
plt.title('Chacha',fontsize=15)
c=['y','b','m','g']
plt.bar(x,y,width=0.2,color='r',label='Popularity')
plt.bar(x,z,width=0.2,color='b',label='Popularity1')
plt.legend()
plt.show()
```

```python
x=['Python','C','c++','Java']
y=[85,60,65,79]
z=[45,36,25,65]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel('Language',fontsize=15)
plt.ylabel('Bhosodi',fontsize=15)
plt.title('Chacha',fontsize=15)

plt.bar(p,y,width,color='r',label='Popularity')
plt.bar(p1,z,width,color='b',label='Popularity1')

plt.xticks(p+width/2,x,rotation=90)
plt.legend()
plt.show()
```

# Horizontal Bar Graph

```python
x=['Python','C','c++','Java']
y=[85,60,65,79]
z=[45,36,25,65]

plt.xlabel('Language',fontsize=15)
plt.ylabel('Bhosodi',fontsize=15)
plt.title('Chacha',fontsize=15)

plt.barh(x,y,color='g',label='Popularity')
plt.barh(x,z,color='y',label='Popularity1')

plt.legend()
plt.show()
```
